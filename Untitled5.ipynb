```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#bar chart about monthly sales
data = pd.DataFrame({
    'Months':['Jan','Feb','Mar','Apr','May','Jun'],
    'sales':[1000,2500,3100,1300,1700,2000]
    })
sns.barplot(x='Months',y='sales',data=data)
plt.title('Sales per Month')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
sns.barplot(x='Months',y='sales',palette='viridis',hue='Months' ,data=data)
plt.title('Sales per Month')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
sns.barplot(x='sales',y='Months',palette='viridis', data=data)
plt.title('Sales per Month')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.show()
```

```python
#changing from bar plot to line plot
sns.lineplot(x='Months',y='sales', data=data, marker='o')
plt.title('Sales per Month')
plt.show()
```

```python
#chatter plot for sales and spending
data = pd.DataFrame({
    'Spending':[2000,1500,3000,2100,1200,1000],
    'sales':[1000,2500,3100,1300,1700,2000]
    })
sns.scatterplot(x='Spending',y='sales',data=data)
plt.title('Sales vs Spending')
plt.show()
```

```python
#Histogram of customer purcahse amounts

purchases = np.random.normal(loc =200, scale=50, size=500)
sns.histplot(purchases, bins=10)
plt.title('Customer Purchase Amounts')
plt.show()
```

```python
purchases = np.random.normal(loc =200, scale=50, size=500)
sns.histplot(purchases, bins=15, color='green')
plt.title('Customer Purchase Amounts')
plt.show()
```

```python
sns.histplot(purchases, bins=15, kde=True, color='green')
plt.title('Customer Purchase Amounts')
plt.show()
```

```python
#boxplot for sales per region
data = pd.DataFrame({
    'Region':['East','East','West','West','North','North']*5,
    'sales':np.random.randint(1000,2500,30)
})
sns.boxplot(x='Region',y='sales',data=data)
plt.title('Sales per Region')
plt.show()
```

```python
sns.boxplot(x='Region',y='sales', palette='viridis', data=data)
plt.title('Sales per Region')
plt.show()
```

```python
#heatmap for correlation

data = pd.DataFrame({
    'sales':np.random.randint(1000,2500,20),
    'profit':np.random.randint(1000,5000,20),
    'spending':np.random.randint(2000,10000,20)
})
sns.heatmap(data.corr())
plt.title('Profit overSales and Spending')
plt.show()
```

```python
sns.heatmap(data.corr(), annot=True)
plt.title('Profit overSales and Spending')
plt.show()
```

```python
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Profit overSales and Spending')
```

```python
#using pairplot for sales, profit, and spending

sns.pairplot(data)
plt.title('Business Reletionship Analytics')
```

```python
#using pairplot for sales, profit, and spending
sns.pairplot(data,diag_kind='kde')
plt.suptitle('Business Relationships Analytics',y=1.1)
plt.show()
```
