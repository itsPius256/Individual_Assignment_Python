# ***Task 2: Data Analysis***

```python
#Counting and printing the unique product categories
unique_categories = df_ecommerce['Category'].nunique() #Calculates the number of unique categories.
print('Product categories:\n',df_ecommerce['Category'].unique())
print(f'Number of unique products categories: {unique_categories}')
```

```python
#Total Number of Transactions
df_ecommerce.shape[0] - 1
```

```python
#Calculates the number of missing values in each column.
df_ecommerce.isnull().sum()
```

```python
# fill missing values
df_ecommerce.loc[df_ecommerce['Price'].isnull() & (df_ecommerce['OrderID'] != 'O00803'), 'Price'] = df_ecommerce['TotalSales'] / df_ecommerce['Quantity'] #Fills missing Price values by calculating them from TotalSales and Quantity.
df_ecommerce.loc[df_ecommerce['Quantity'].isnull() & (df_ecommerce['OrderID'] != 'O00803'), 'Quantity'] = df_ecommerce['TotalSales'] / df_ecommerce['Price'] #Fills missing Quantity values by calculating them from TotalSales and Price.
display(df_ecommerce.isnull().sum())#Displays updated missing values count to ensure most are handled
```

```python
# filling missing values for order O00803
df_ecommerce.loc[df_ecommerce['OrderID'] == 'O00803', 'Price'] = 22 #Fills the missing Price value for a specific order.
df_ecommerce.loc[df_ecommerce['OrderID'] == 'O00803', 'Quantity'] = 1 #Fills the missing Quantity value for a specific order.
df_ecommerce.isnull().sum()#Displays updated missing values count to ensure most are handled
```

```python
#negative totalsales
neg_total_sales = df_ecommerce[df_ecommerce['TotalSales'] < 0] #Filtering rows with negative total sales
print(neg_total_sales)

for column in df_ecommerce.select_dtypes(include=['number']).columns: #Loops through numeric columns to find which ones have negative values
  neg_count = (df_ecommerce[column] < 0).sum() #Calculates the number of negative values in each column.

  if neg_count > 0: #Checks if there are any negative values.
    print(f"- {column}: {neg_count} negative value(s)")

df_ecommerce.head()
```

```python
#replacing the negative total sales by the product of the price and quantity
df_ecommerce.loc[df_ecommerce['TotalSales'] < 0, 'TotalSales'] = df_ecommerce['Price'] * df_ecommerce['Quantity']
```

```python
#show the negative total sales in the dataset
neg_total_sales = df_ecommerce[df_ecommerce['TotalSales'] < 0]
print(neg_total_sales)
```

```python
#Saved the cleaned dataframe to a new CSV file named 'df1_ecommerce.csv' without the index column.
df_ecommerce.to_csv('/content/df1_ecommerce.csv', index=False)
df1_ecommerce = pd.read_csv('/content/df1_ecommerce.csv') #Read the newly created CSV file into a new dataframe called 'df1_ecommerce'.
df1_ecommerce.head()
```

# Task 3: ***Visualization***

```python
#importing the libraries
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set(style='whitegrid') #Sets default style to clean and readable grid layout.
os.makedirs('graph', exist_ok=True) #Creating a folder named graph to save the plots
```

```python
#visualization
#The product category with the most sales volume

top_categories = df1_ecommerce.groupby('Category')['TotalSales'].sum().sort_values(ascending=False) #Groups data by category and calculates the total sales for each, then sorts them in descending order.

plt.figure(figsize=(10,6)) #Sets the figure size for the plot.
sns.barplot(x=top_categories.index, y=top_categories.values,palette='viridis') #Creates a bar plot of the top categories.
plt.title('Top 10 Product Categories by TotalSales')
plt.xlabel('TotalSales')
plt.ylabel('Category')
plt.tight_layout() #Adjusts plot to ensure everything fits without overlapping.
plt.savefig('graph/task3_top10_categories.png')
plt.show()
```

```python
#The year with the most number of orders
df1_ecommerce['OrderDate'] = pd.to_datetime(df1_ecommerce['OrderDate']) #Converting OrderDate to datetime and extract the year
df1_ecommerce['Order Year'] = df1_ecommerce['OrderDate'].dt.year #Extracted the year from the OrderDate.

orders_per_year = df1_ecommerce['Order Year'].value_counts().sort_index() #Counts the number of orders per year and sorts by year.

plt.figure(figsize=(8,5))
sns.barplot(x=orders_per_year.index, y=orders_per_year.values, hue=orders_per_year.index, palette='inferno', legend=False)
plt.title('Number of Orders by Year')
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.savefig('graph/task4_orders_per_year.png')
plt.show()
```

```python
#Creating a bar graph for the countries with the highest total sales
from matplotlib.ticker import FuncFormatter

def thousands_formatter(x, pos): #formating numbers in thousands.
    return f'{x/1000:.0f}k'

revenue_by_country = df1_ecommerce.groupby('Country')['TotalSales'].sum().sort_values(ascending=False)# Groups data by country, calculate total sales per country and rank the  countries from highest to lowest sales.

plt.figure(figsize=(10,6))#Set the size of the plot to 10x6 inches
ax = sns.barplot(x=revenue_by_country.values, y=revenue_by_country.index, palette='magma')#Draws a horizontal barplot, X = revenue values, Y = country names,'magma' to give a fiery color scheme to visually emphasize high values.
ax.xaxis.set_major_formatter(FuncFormatter(thousands_formatter))
plt.title('Top countries by Total Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Country')
for i, v in enumerate(revenue_by_country.values):#Annotates each bar with its value, adds 100 offset so the text does not overlap with the bar and  adds commas and 2 decimal points for better readability
  plt.text(v + 100, i, f'{v:,.2f}', va= 'center')
plt.tight_layout() # to aviod label overlap
plt.savefig('graph/task5_top_countries_by_revenue.png')
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter

def thousands_formatter(x, pos):
    return f'{x/1000:.0f}k'

revenue_df = revenue_by_country.reset_index() #converting the revenue by country into a dataframe
revenue_df.columns = ['Country', 'Total Revenue']#rename the columnsto make the labeling clear

ax= sns.barplot(data= revenue_df,x = 'Total Revenue', y = 'Country', palette='Blues_d')
ax.xaxis.set_major_formatter(FuncFormatter(thousands_formatter)) #Applies the thousands formatter to the x-axis.
plt.title('Top countries by Total Revenue (Improved)')
plt.xlabel('Total Revenue')
plt.ylabel('Country')

for i, v in enumerate(revenue_by_country.values):#Annotates each bar with its value, adds 100 offset so the text does not overlap with the bar and  adds commas and 2 decimal points for better readability
  plt.text(v + 100, i, f'{v:,.2f}', va= 'center')

plt.tight_layout()
plt.savefig('graph/task4_top_countries_by_revenue_improved.png')
plt.show()
```

```python
#Calculating the standard deviation of total sales for each country and sorts them.
sales_std = df1_ecommerce.groupby('Country')['TotalSales'].std().sort_values()

most_consistent= sales_std.head(5) #Gets the 5 countries with the most consistent sales which have the lowest standard deviation
print(most_consistent)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=most_consistent.values, y=most_consistent.index, hue=most_consistent.index, palette='coolwarm', legend=False)
plt.title('Countries with the Most Consistent Sales')
plt.xlabel('Standard Deviation of TotalSales')

for i, v in enumerate(most_consistent.values): #Annotates each bar with its value.
  plt.text(v + 100, i, f'{v:,.2f}', va= 'center')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig('graph/task7_most_consistent_sales.png')
plt.show()
```

```python
# Calculate mean and standard deviation of TotalSales for the entire dataset using a histogram
mean_sales = df1_ecommerce['TotalSales'].mean()
std_sales = df1_ecommerce['TotalSales'].std()

plt.figure(figsize=(10, 6))
sns.histplot(data=df1_ecommerce, x='TotalSales', bins=30, kde=True) #a histogram of TotalSales with 30 bins and adds a Kernel Density Estimate (KDE) line.
plt.axvline(mean_sales, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_sales:.2f}') #Adds a vertical dashed red line at the mean of TotalSales.
plt.axvline(mean_sales + std_sales, color='green', linestyle='dashed', linewidth=2, label=f'+1 Std Dev: {mean_sales + std_sales:.2f}') #Add a vertical dashed green line at one standard deviation above the mean.
plt.axvline(mean_sales - std_sales, color='orange', linestyle='dashed', linewidth=2, label=f'-1 Std Dev: {mean_sales - std_sales:.2f}') #And also a vertical dashed orange line at one standard deviation below the mean.

plt.title('Distribution of Total Sales with Mean and Standard Deviation')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.legend() #Display the legend for the mean and standard deviation lines.
plt.tight_layout()
plt.show()
```

```python
# Converting categorical variables to numerical using one-hot encoding
df_encoded = pd.get_dummies(df1_ecommerce, columns=['Category', 'Country', 'PaymentMethod', 'CustomerSegment'])
```

```python
# Select 'TotalSales' and all one-hot encoded category columns
category_sales_correlation_df = df_encoded[['TotalSales'] + [col for col in df_encoded.columns if col.startswith('Category_')]]

# Calculate the correlation matrix
correlation_matrix_category_sales = category_sales_correlation_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix_category_sales, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap: Category vs Total Sales')
plt.tight_layout()
plt.show()
```
