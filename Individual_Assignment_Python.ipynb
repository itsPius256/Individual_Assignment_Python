```python
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set(style='whitegrid')
os.makedirs('graph', exist_ok=True)
```

```python
top_categories = df_ecommerce.groupby('Category')['TotalSales'].sum().sort_values(ascending=False).head()

plt.figure(figsize=(10,6))
sns.barplot(x=top_categories.index, y=top_categories.values,palette='viridis')
plt.title('Top 10 Product Categories by TotalSales')
plt.xlabel('TotalSales')
plt.ylabel('Category')
plt.tight_layout()
plt.savefig('graph/task3_top10_categories.png')
plt.show()
```

```python
df_ecommerce['OrderDate'] = pd.to_datetime(df_ecommerce['OrderDate'])
df_ecommerce['Order Year'] = df_ecommerce['OrderDate'].dt.year

orders_per_year = df_ecommerce['Order Year'].value_counts().sort_index()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x=orders_per_year.index, y=orders_per_year.values, hue=orders_per_year.index, palette='cubehelix', legend=False)
plt.title('Number of Orders by Year')
plt.xlabel('Year')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.savefig('graph/task4_orders_per_year.png')
plt.show()
```

```python
revenue_by_country = df_ecommerce.groupby('Country')['TotalSales'].sum().sort_values(ascending=False).head()

plt.figure(figsize=(10,6))
sns.barplot(x=revenue_by_country.values, y=revenue_by_country.index, palette='magma')
plt.title('Top countries by Total Revenue')
plt.xlabel('Total Revenue')
plt.ylabel('Country')
for i, v in enumerate(revenue_by_country.values):
  plt.text(v + 100, i, f'{v:,.2f}', va= 'center')
plt.tight_layout()
plt.savefig('graph/task5_top_countries_by_revenue.png')
plt.show()
```

```python
revenue_df = revenue_by_country.reset_index()
revenue_df.columns = ['Country', 'Total Revenue']

ax= sns.barplot(data= revenue_df,x = 'Total Revenue', y = 'Country', palette='Blues_d')
plt.title('Top countries by Total Revenue (Improved)')
plt.xlabel('Total Revenue')
plt.ylabel('Country')

for i, v in enumerate(revenue_by_country.values):
  plt.text(v + 100, i, f'{v:,.2f}', va= 'center')

plt.tight_layout()
plt.savefig('graph/task4_top_countries_by_revenue_improved.png')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_usa_products.head(20).values, y=top_usa_products.head(20).index, hue=top_usa_products.head(20).index, palette='pastel', legend=False)
plt.title('Top 20 Best-selling Products in the United States')
plt.xlabel('Total Revenue')
plt.ylabel('Product')
plt.tight_layout()
plt.savefig('graph/Top_20_Best_-_selling_Products_in_the_United_States.png')
plt.show()
```

```python
sales_std = df_ecommerce.groupby('Country')['TotalSales'].std().sort_values()

most_consistent= sales_std.head(5)
print(most_consistent)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=most_consistent.values, y=most_consistent.index, hue=most_consistent.index, palette='coolwarm', legend=False)
plt.title('Countries with the Most Consistent Sales')
plt.xlabel('Standard Deviation of TotalSales')

for i, v in enumerate(most_consistent.values):
  plt.text(v + 100, i, f'{v:,.2f}', va= 'center')
plt.ylabel('Country')
plt.tight_layout()
plt.savefig('graph/task7_most_consistent_sales.png')
plt.show()
```

```python
df_ecommerce['Revenue'] = df_ecommerce['Price'] * df_ecommerce['Quantity']
numerical_df_ecommerce = df_ecommerce.select_dtypes(include=['float'])

correlation_matrix = numerical_df_ecommerce.corr()
print('correlation matrix:')
print(correlation_matrix)
```

```python
category_revenue = df_ecommerce.pivot_table(values ='Revenue', index='Category', aggfunc='sum')
plt.figure(figsize=(10,8))
sns.heatmap(category_revenue, annot=True, fmt='.2f',cmap='YlGnBu')
plt.title('Total Revenue by Product Category')
plt.xlabel('Revenue')
plt.ylabel('Category')
plt.tight_layout()
plt.savefig('graph/task7_revenue_by_product_category.png')
plt.show()
```
